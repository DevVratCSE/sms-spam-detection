```python
# ===============================================================
# Project: Handwritten Digit Recognition using MNIST
# ===============================================================
#
# Objective:
# The goal of this project is to build an intelligent system capable
# of recognizing handwritten digits (0–9) from grayscale images.
# This project demonstrates fundamental machine learning and deep
# learning techniques used in modern computer vision.
#
# What I am trying to accomplish:
# • Understand the complete ML workflow — data loading, preprocessing,
#   model training, evaluation, and visualization.
# • Compare classical machine learning with deep learning approaches.
# • Optimize training performance using GPU acceleration.
#
# Models Implemented:
#
# 1️⃣ Classical Machine Learning Baseline
#    - Logistic Regression trained on flattened pixel values.
#    - Serves as a performance benchmark.
#
# 2️⃣ Deep Learning Models
#
#    ✔ Multi-Layer Perceptron (MLP)
#      - Fully connected neural network.
#      - Learns complex nonlinear relationships in image data.
#
#    ✔ Convolutional Neural Network (CNN)
#      - Designed specifically for image recognition tasks.
#      - Automatically extracts spatial features such as edges,
#        curves, and shapes.
#      - Expected to outperform classical methods.
#
# Performance Optimization:
# • Google Colab T4 GPU acceleration
# • Mixed Precision (FP16) for faster computation
# • XLA JIT compilation
# • Efficient tf.data pipelines (cache + prefetch)
# • Early stopping to prevent overfitting
#
# Evaluation Methods:
# • Accuracy comparison across models
# • Confusion matrices
# • Classification report
# • Training vs validation curves
# • Visualization of predicted vs actual digits
#
# Expected Outcome:
# The CNN model should achieve the highest accuracy (~99%),
# demonstrating the power of deep learning for image classification.
#
# Skills Demonstrated:
# Python | TensorFlow | Keras | Scikit-learn
# Deep Learning | Computer Vision | Model Evaluation
# GPU Training | Data Pipelines | Performance Tuning
#
# ===============================================================

# ===========================
# MNIST: Baseline (LogReg) + Deep Learning (MLP + CNN)
# Single-cell, Colab T4-GPU optimized (tf.data + mixed precision + XLA)
# ===========================

# --- Install deps (quiet) ---
!pip -q install scikit-learn matplotlib seaborn

import os, numpy as np, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf

from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ---------------------------
# 0) Speed / GPU Settings
# ---------------------------
print("TensorFlow:", tf.__version__)
print("Physical GPUs:", tf.config.list_physical_devices("GPU"))

gpus = tf.config.list_physical_devices("GPU")
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("✅ GPU memory growth enabled")
    except Exception as e:
        print("⚠️ Could not set memory growth:", e)

# XLA JIT (often faster)
try:
    tf.config.optimizer.set_jit(True)
    print("✅ XLA JIT enabled")
except Exception as e:
    print("⚠️ XLA enable failed:", e)

# Mixed precision (best on T4). Keep final softmax in float32.
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy("mixed_float16")
print("✅ Mixed precision policy:", mixed_precision.global_policy())

# ---------------------------
# 1) Load MNIST
# ---------------------------
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print("Train:", X_train.shape, y_train.shape)
print("Test :", X_test.shape, y_test.shape)

# Preview digits
plt.figure(figsize=(6,2))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(X_train[i], cmap="gray")
    plt.axis("off")
    plt.title(str(y_train[i]))
plt.tight_layout()
plt.show()

# ---------------------------
# 2) Baseline: Logistic Regression (sklearn) - CPU-bound but quick
# ---------------------------
X_train_flat_full = X_train.reshape(-1, 784).astype("float32") / 255.0
X_test_flat = X_test.reshape(-1, 784).astype("float32") / 255.0

logreg = LogisticRegression(max_iter=200, n_jobs=-1)
logreg.fit(X_train_flat_full, y_train)

y_pred_lr = logreg.predict(X_test_flat)
acc_lr = accuracy_score(y_test, y_pred_lr)

print("\n=== Baseline: Logistic Regression ===")
print("Accuracy:", acc_lr)
print(classification_report(y_test, y_pred_lr))

cm_lr = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(8,6))
sns.heatmap(cm_lr, annot=False, cmap="Blues", fmt="d")
plt.title("Confusion Matrix — Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ---------------------------
# 3) Deep Learning Prep (tf.data)
# ---------------------------
# Normalize + reshape for CNN
X_train_cnn = (X_train.astype("float32") / 255.0)[..., None]  # (N, 28,28,1)
X_test_cnn  = (X_test.astype("float32")  / 255.0)[..., None]

y_train_oh = to_categorical(y_train, 10)
y_test_oh  = to_categorical(y_test, 10)

# Train/Val split (simple fixed split)
val_size = 6000
X_val_cnn, y_val_oh = X_train_cnn[:val_size], y_train_oh[:val_size]
X_train_cnn2, y_train_oh2 = X_train_cnn[val_size:], y_train_oh[val_size:]

BATCH_SIZE = 256  # tuned for T4; try 512 if you want more speed and no OOM
AUTOTUNE = tf.data.AUTOTUNE

def make_ds(X, y, training=False):
    ds = tf.data.Dataset.from_tensor_slices((X, y))
    if training:
        ds = ds.shuffle(20000, reshuffle_each_iteration=True)
    ds = ds.cache()  # MNIST fits in RAM
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(X_train_cnn2, y_train_oh2, training=True)
val_ds   = make_ds(X_val_cnn, y_val_oh, training=False)
test_ds  = make_ds(X_test_cnn, y_test_oh, training=False)

# Also prepare flattened data for MLP (uses same split)
X_train_mlp2 = X_train_cnn2.reshape(-1, 784)
X_val_mlp    = X_val_cnn.reshape(-1, 784)
X_test_mlp   = X_test_cnn.reshape(-1, 784)

# ---------------------------
# 4) Deep Learning Model A: MLP
# ---------------------------
def build_mlp():
    m = models.Sequential([
        layers.Input(shape=(784,)),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.2),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.2),
        layers.Dense(10, activation="softmax", dtype="float32")  # keep output stable
    ])
    m.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return m

mlp = build_mlp()
print("\n=== Training MLP ===")
mlp_cb = [
    callbacks.EarlyStopping(monitor="val_accuracy", patience=2, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=1, min_lr=1e-5),
]

hist_mlp = mlp.fit(
    X_train_mlp2, y_train_oh2,
    validation_data=(X_val_mlp, y_val_oh),
    epochs=10,
    batch_size=BATCH_SIZE,
    callbacks=mlp_cb,
    verbose=1
)

mlp_loss, mlp_acc = mlp.evaluate(X_test_mlp, y_test_oh, verbose=0)
print("✅ MLP Test Accuracy:", float(mlp_acc))

plt.figure()
plt.plot(hist_mlp.history["accuracy"], label="train_acc")
plt.plot(hist_mlp.history["val_accuracy"], label="val_acc")
plt.title("MLP Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.figure()
plt.plot(hist_mlp.history["loss"], label="train_loss")
plt.plot(hist_mlp.history["val_loss"], label="val_loss")
plt.title("MLP Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# ---------------------------
# 5) Deep Learning Model B: CNN (best fit)
# ---------------------------
def build_cnn():
    m = models.Sequential([
        layers.Input(shape=(28,28,1)),
        layers.Conv2D(32, (3,3), activation="relu"),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation="relu"),
        layers.MaxPooling2D((2,2)),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(10, activation="softmax", dtype="float32")  # output float32 for stability
    ])
    m.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return m

cnn = build_cnn()
print("\n=== Training CNN ===")
cnn_cb = [
    callbacks.EarlyStopping(monitor="val_accuracy", patience=2, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=1, min_lr=1e-5),
]

hist_cnn = cnn.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=cnn_cb,
    verbose=1
)

cnn_loss, cnn_acc = cnn.evaluate(test_ds, verbose=0)
print("✅ CNN Test Accuracy:", float(cnn_acc))

plt.figure()
plt.plot(hist_cnn.history["accuracy"], label="train_acc")
plt.plot(hist_cnn.history["val_accuracy"], label="val_acc")
plt.title("CNN Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.figure()
plt.plot(hist_cnn.history["loss"], label="train_loss")
plt.plot(hist_cnn.history["val_loss"], label="val_loss")
plt.title("CNN Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# ---------------------------
# 6) CNN Confusion Matrix + Predicted vs Actual
# ---------------------------
y_pred_cnn = np.argmax(cnn.predict(test_ds, verbose=0), axis=1)

cm_cnn = confusion_matrix(y_test, y_pred_cnn)
plt.figure(figsize=(8,6))
sns.heatmap(cm_cnn, annot=False, cmap="Blues", fmt="d")
plt.title("Confusion Matrix — CNN")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Show random predictions
idx = np.random.choice(len(X_test), 20, replace=False)
plt.figure(figsize=(12,4))
for i, j in enumerate(idx):
    plt.subplot(2,10,i+1)
    plt.imshow(X_test[j], cmap="gray")
    plt.axis("off")
    plt.title(f"T:{y_test[j]}\nP:{y_pred_cnn[j]}")
plt.tight_layout()
plt.show()

# ---------------------------
# 7) Final Summary
# ---------------------------
print("\n=== Model Comparison Summary ===")
print(f"LogReg Accuracy: {acc_lr:.4f}")
print(f"MLP   Accuracy: {mlp_acc:.4f}")
print(f"CNN   Accuracy: {cnn_acc:.4f}")

```

# Working Mechanism & Project Summary

### 1. Working Mechanism
This project implements a multi-stage pipeline for handwritten digit recognition using the MNIST dataset.

*   **Data Acquisition & Preprocessing**:
    *   Loads 70,000 grayscale images ($28 \times 28$ pixels).
    *   **Normalization**: Pixel values are scaled from $[0, 255]$ to $[0, 1]$ for numerical stability.
    *   **Reshaping**: Data is transformed into flattened vectors for LogReg/MLP and 4D tensors $(N, 28, 28, 1)$ for CNN.
    *   **Pipelines**: Uses `tf.data` with caching and prefetching to ensure the GPU is never starved of data.

*   **Model Architectures**:
    *   **Logistic Regression (Baseline)**: A simple linear model using Scikit-Learn. It sets a benchmark of ~92.6%.
    *   **Multi-Layer Perceptron (MLP)**: A deep neural network with 2 hidden layers (256 and 128 neurons). It uses Dropout to prevent overfitting and reaches ~98% accuracy.
    *   **Convolutional Neural Network (CNN)**: The state-of-the-art approach. It uses spatial filters to detect edges and patterns, achieving the highest accuracy (~99%).

*   **Performance Optimization**:
    *   **Mixed Precision**: Uses `float16` for faster computation on the T4 GPU while keeping the output layer in `float32`.
    *   **XLA Compilation**: Just-In-Time (JIT) compilation to optimize the TensorFlow graph.

### 2. Summary of Findings
| Model | Accuracy | Strengths |
| :--- | :--- | :--- |
| **LogReg** | ~92.6% | Simple, fast training, low computational cost. |
| **MLP** | ~98.1% | Captures non-linearities better than LogReg. |
| **CNN** | ~99.0% | Superior feature extraction (spatial awareness). |

### 3. Key Insights
*   **Data Distribution**: The model performs best when real-world inputs (like sketches) are preprocessed to match the MNIST distribution (centered, $28 \times 28$, and white-on-black).
*   **Infrastructure**: Leveraging GPU acceleration and optimized data pipelines allows for training sophisticated models in seconds rather than minutes.
